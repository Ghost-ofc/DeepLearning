# Curacion de dataset

Se realizó una curación del dataset utilizado para el modelo en cuestión, ya que es un paso esencial en el proceso de preparación de datos para el entrenamiento. Durante los entrenamientos, se llegó a un punto en el que el modelo aprendió a asociar que lo que era verde correspondía a "hojas", lo que ocasionaba que confundiera ranas con hojas. Para solucionar esto, se llevó a cabo una curación exhaustiva del dataset, eliminando imágenes no representativas o con características ambiguas, lo que permitió mejorar el rendimiento del modelo y reducir los errores de clasificación.

## Modelo 1

**Distribuciones de brillo y contraste para los datasets antes de la curacion**



- La distribución del brillo muestra un mayor desajuste en las frecuencias de las clases, con algunas clases mostrando picos y otros sin una distribución uniforme, lo que sugiere que puede haber datos ruidosos o inconsistentes.

- Las clases parecen estar dispersas de manera desigual en las diferentes intensidades de brillo, lo cual puede ser un indicio de que los datos no están equilibrados o bien organizados.

- El contraste muestra una gran variabilidad entre las clases. Se pueden observar algunas frecuencias de clase muy dispersas, lo que también indica la presencia de datos erróneos, posibles errores de anotación o falta de consistencia en la forma en que se recogieron los datos.

- Algunas clases tienen picos muy altos en contraste, mientras que otras apenas muestran presencia, lo cual puede afectar negativamente el desempeño del modelo si se alimenta de datos desequilibrados.

**Despues de la curacion del dataset**



- Después de la curación del dataset, en la segunda gráfica se observa una distribución mucho más uniforme entre las clases en cuanto al brillo. Las clases parecen estar equilibradas y las distribuciones son más consistentes.

- Esto indica que la curación del dataset ha ayudado a eliminar datos erróneos o ruidosos, lo cual facilita el análisis y mejora la calidad de los datos para el modelo de aprendizaje automático.

- Después de la curación, la distribución del contraste es más equilibrada y consistente. Las clases están mucho mejor distribuidas a lo largo de las intensidades de contraste, lo que sugiere que la curación ha logrado hacer los datos más representativos y homogéneos.

- Esto permite que el modelo de aprendizaje automático entienda mejor los patrones de las diferentes clases, mejorando la precisión y reduciendo los posibles errores de clasificación.
