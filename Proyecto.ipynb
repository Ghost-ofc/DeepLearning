# Deteccion de Plagas Mediante CNN y Transfer Learning

## Problematica

¿Cómo la implementación de redes neuronales convolucionales (CNN) con transferencia de aprendizaje en una aplicación móvil puede optimizar la detección temprana de roya (Puccinia spp.) y mosca minadora (Liriomyza spp.) en cultivos agrícolas, superando las limitaciones de los métodos tradicionales de inspección visual?

Estas plagas generan daños económicos significativos: la roya afecta el follaje reduciendo la capacidad fotosintética (Castellano-Corzo et al., 2023), mientras la mosca minadora produce galerías en hojas que debilitan las plantas (Ferrer Martínez, 2021). Los métodos convencionales presentan tres limitaciones principales: 1) Requieren expertise agronómico especializado, 2) Tienen alta variabilidad en los diagnósticos, y 3) Resultan ineficientes para grandes extensiones de cultivo (Islam et al., 2019; Repositorio UTMachala, 2022).

La propuesta se sustenta en evidencias recientes:

- Las CNN logran hasta 87% de precisión en detección de roya usando imágenes móviles (Castellano-Corzo et al., 2023)
- El transfer learning permite alcanzar 96.65% de exactitud con datasets limitados (Ferrer Martínez, 2021)
- Modelos híbridos CNN-transfer learning muestran 90.8% de efectividad en detección de plagas (Ballena Ruiz & Rivera Cartagena, 2022)
- Implementaciones móviles usando SqueezeNet reducen 99% los parámetros vs arquitecturas tradicionales (Ferrer Martínez, 2021)

## Hipotesis

La implementación de redes neuronales convolucionales (CNN) con transferencia de aprendizaje en una aplicación móvil puede optimizar la detección temprana de roya (Puccinia spp.) y mosca minadora (Liriomyza spp.) en cultivos agrícolas, superando las limitaciones de los métodos tradicionales de inspección visual, al ofrecer diagnósticos más precisos, rápidos y accesibles, incluso en grandes extensiones de cultivo, lo que resultará en una mejora significativa en la eficiencia y efectividad del control de plagas, con una tasa de precisión superior al 87% en comparación con los métodos convencionales.

## Datos de entrada

- **Imágenes de hojas con plaga:** Imágenes de hojas afectadas, mostrando los diferentes daños causados por la plaga.
